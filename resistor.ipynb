```python
import pandas as pd
import os
import shutil

# İşlenecek klasörler
base_path = "/content/resistor_classification"  # kendi yolunla değiştir
splits = ['valid', 'test']

for split in splits:
    csv_path = os.path.join(base_path, split, "_classes.csv")
    if not os.path.exists(csv_path):
        print(f"Uyarı: {split} klasöründe _classes.csv bulunamadı, atlanıyor.")
        continue

    df = pd.read_csv(csv_path)
    df.columns = df.columns.str.strip()  # Başlıkları düzelt
    filenames = df.iloc[:, 0].values
    class_names = df.columns[1:]  # Sınıf isimleri başlıklarda

    for idx, row in df.iterrows():
        filename = row[0]
        label_index = row[1:].astype(int).idxmax()  # en yüksek olan 1 değerli kolon
        label_name = label_index.strip()

        src_file = os.path.join(base_path, split, filename)
        dst_dir = os.path.join(base_path, split, label_name)

        os.makedirs(dst_dir, exist_ok=True)

        if os.path.exists(src_file):
            shutil.move(src_file, os.path.join(dst_dir, filename))
        else:
            print(f"Dosya bulunamadı: {src_file}")
```
