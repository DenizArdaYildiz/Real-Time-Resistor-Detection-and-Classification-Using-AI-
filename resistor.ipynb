```python
import pandas as pd
import os
import shutil

# Ä°ÅŸlenecek klasÃ¶rler
base_path = "/content/resistor_classification"  # kendi yolunla deÄŸiÅŸtir
splits = ['valid', 'test']

for split in splits:
    csv_path = os.path.join(base_path, split, "_classes.csv")
    if not os.path.exists(csv_path):
        print(f"UyarÄ±: {split} klasÃ¶rÃ¼nde _classes.csv bulunamadÄ±, atlanÄ±yor.")
        continue

    df = pd.read_csv(csv_path)
    df.columns = df.columns.str.strip()  # BaÅŸlÄ±klarÄ± dÃ¼zelt
    filenames = df.iloc[:, 0].values
    class_names = df.columns[1:]  # SÄ±nÄ±f isimleri baÅŸlÄ±klarda

    for idx, row in df.iterrows():
        filename = row[0]
        label_index = row[1:].astype(int).idxmax()  # en yÃ¼ksek olan 1 deÄŸerli kolon
        label_name = label_index.strip()

        src_file = os.path.join(base_path, split, filename)
        dst_dir = os.path.join(base_path, split, label_name)

        os.makedirs(dst_dir, exist_ok=True)

        if os.path.exists(src_file):
            shutil.move(src_file, os.path.join(dst_dir, filename))
        else:
            print(f"Dosya bulunamadÄ±: {src_file}")
```
